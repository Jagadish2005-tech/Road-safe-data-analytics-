```python
df.drop(['End_Lat', 'End_Lng'], axis=1, inplace=True)  # Drop the columns of 'End_lat' & 'End_lng' from this dataset.
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer

# Select only numeric columns
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
numeric_df = df[numeric_cols]

# Step 1: Create SimpleImputer (mean strategy, you can also use 'median' or 'most_frequent')
imputer = SimpleImputer(strategy="mean")

# Step 2: Fit and transform the numeric data
imputed_array = imputer.fit_transform(numeric_df)

# Step 3: Convert back to DataFrame
imputed_df = pd.DataFrame(imputed_array, columns=numeric_cols, index=df.index)

# Step 4: Replace numeric columns in original df
df[numeric_cols] = imputed_df

print("Missing values imputed successfully using mean strategy.")
print(df[numeric_cols].isnull().sum())
```

```python
from sklearn.impute import SimpleImputer

# Select categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns
categorical_df = df[categorical_cols]

# Impute with most frequent value
cat_imputer = SimpleImputer(strategy="most_frequent")           # for categorical values
imputed_cat = cat_imputer.fit_transform(categorical_df)

# Convert back to DataFrame
imputed_cat_df = pd.DataFrame(imputed_cat, columns=categorical_cols, index=df.index)

# Replace in original df
df[categorical_cols] = imputed_cat_df

print("Missing values in categorical columns imputed with mode.")
print(df[categorical_cols].isnull().sum())
```
