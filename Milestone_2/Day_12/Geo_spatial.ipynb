```python
import pandas as pd

# Load cleaned dataset
df = pd.read_csv("US_Accidents_Cleaned.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assume df is your cleaned accident dataset with 'Start_Lat' and 'Start_Lng'
# Drop rows with missing coordinates
geo_df = df.dropna(subset=['Start_Lat', 'Start_Lng'])

# Optional: Sample large datasets for faster plotting
sample_df = geo_df.sample(n=100000, random_state=42)  # Adjust n for performance

# Scatter plot of accident locations
plt.figure(figsize=(12, 8))
plt.scatter(sample_df['Start_Lng'], sample_df['Start_Lat'], 
            alpha=0.1, s=5, c='red')
plt.title('Accident Locations Scatter Plot (Sampled)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
1.Heatmap with Seaborn for density
```

```python
plt.figure(figsize=(12, 8))
plt.hexbin(
    geo_df['Start_Lng'],
    geo_df['Start_Lat'],
    gridsize=200, cmap='Reds', mincnt=1
)
plt.colorbar(label='Accident Count')
plt.title('Accident Density Hexbin Plot')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Drop rows with missing coordinates
geo_df = df.dropna(subset=['Start_Lat', 'Start_Lng'])

# Sample for performance (adjust n if needed)
sample_df = geo_df.sample(n=500000, random_state=42)  # 500k points

# Scatter plot
plt.figure(figsize=(12, 8))
plt.scatter(
    sample_df['Start_Lng'],
    sample_df['Start_Lat'],
    s=5,           # size of points
    alpha=0.2,     # transparency for overlapping points
    c='red'        # color
)
plt.title('Accident Locations Scatter Plot (Sampled)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
